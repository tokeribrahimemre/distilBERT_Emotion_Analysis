#Pull all Comments for Multiple Videos

## 1. Get Comments Function

```python
import googleapiclient.discovery
import pandas as pd

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = dev

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)


def getcomments(video):
  request = youtube.commentThreads().list(
      part="snippet",
      videoId=video,
      maxResults=100
  )

  comments = []

  # İsteği çalıştır
  response = request.execute()

  # Yanıtın içindeki yorumları alın.
  for item in response['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      public = item['snippet']['isPublic']
      comments.append([
          comment['authorDisplayName'],
          comment['publishedAt'],
          comment['likeCount'],
          comment['textOriginal'],
          comment['videoId'],
          public
      ])

  while (1 == 1):
    try:
     nextPageToken = response['nextPageToken']
    except KeyError:
     break
    nextPageToken = response['nextPageToken']
    # Bir sonraki sayfa belirteci ile yeni bir istek nesnesi oluşturun.
    nextRequest = youtube.commentThreads().list(part="snippet", videoId=video, maxResults=100, pageToken=nextPageToken)
    # Sonraki isteği çalıştır.
    response = nextRequest.execute()
    # Bir sonraki yanıtın içindeki yorumları alın.
    for item in response['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      public = item['snippet']['isPublic']
      comments.append([
          comment['authorDisplayName'],
          comment['publishedAt'],
          comment['likeCount'],
          comment['textOriginal'],
          comment['videoId'],
          public
      ])

  df2 = pd.DataFrame(comments, columns=['author', 'updated_at', 'like_count', 'text','video_id','public'])
  return df2
```

## 2. For Loop for List of IDs

```python
df = pd.DataFrame()
for i in ['2fAPgOCjToA','anmuklFtu8U','NXpdyAWLDas','rWpiYvLHmzU','tWP6z0hvw1M','XllmgXBQUwA','q05kiN7P-eY','SkCz4LqzNF4']:
  df2 = getcomments(i)
  df = pd.concat([df, df2])
```

```python
import pandas as pd
import re
from langdetect import detect
from nltk.tokenize import word_tokenize
import nltk
import emoji

# Gerekli paketleri yükleyin
!pip install langdetect emoji

# Gerekli NLTK veri dosyalarını indirin
nltk.download('words')
nltk.download('punkt')

# CSV dosyasını okuyun
file_path = "/content/drive/MyDrive/Colab Notebooks/YouTube_Data/soraAi_data.csv"
data = pd.read_csv(file_path)
```

```python
from nltk.corpus import words

# İngilizce kelimeleri nltk'den çekin
english_words = set(words.words())

# Emojileri kaldırmak için daha kapsamlı bir regex deseni tanımlayın
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F1E0-\U0001F1FF"  # Flags (iOS)
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed Characters
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002500-\U00002BEF"  # Box Drawing
        "\U0001F700-\U0001F77F"  # Alchemical Symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# URL'leri kaldırmak için bir fonksiyon tanımlayın
def remove_urls(text):
    return re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

# İngilizce olmayan satırları tespit edip çıkarmak için bir fonksiyon tanımlayın
def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

# İngilizce olmayan kelimeleri temizlemek için bir fonksiyon tanımlayın
def remove_non_english_words(text, english_words):
    words = text.split()
    cleaned_words = [word for word in words if word.lower() in english_words]
    return ' '.join(cleaned_words)

# Metin temizleme işlemi için bir fonksiyon tanımlayın
def clean_text(text):
    text = str(text)
    text = remove_urls(text)  # URL'leri kaldırma
    text = remove_emojis(text)  # Emojileri kaldırma
    text = re.sub(r'\W', ' ', text)  # Özel karakterleri kaldırma
    text = re.sub(r'\s+', ' ', text)  # Gereksiz boşlukları kaldırma
    text = text.lower()  # Küçük harfe dönüştürme
    text = remove_non_english_words(text, english_words)  # İngilizce olmayan kelimeleri kaldırma
    words = word_tokenize(text)  # Kelimelere ayırma
    return ' '.join(words)

# Verideki tüm metin sütunlarına temizleme işlemi uygulayın
for column in data.select_dtypes(include=[object]).columns:
    data = data[data[column].apply(is_english)]  # İngilizce olmayan satırları çıkarma
    data[column] = data[column].apply(clean_text)
    # 3 kelimeden az satırları çıkarma
    data = data[data[column].apply(lambda x: len(x.split()) >= 3)]

# Eksik verileri çıkarma
data = data.dropna()

# Temizlenmiş veriyi yeni bir CSV dosyasına kaydedin
output_file_path = '/mnt/data/sora_cleaned.csv'
data.to_csv(output_file_path, index=False)

print("Veri temizleme işlemi tamamlandı ve temizlenmiş veri kaydedildi.")
```
